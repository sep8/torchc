```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
w = 1.0
```

```python
def forward(x):
    return x * w
```

```python
def cost(xs, ys):
    cost = 0
    for x, y in zip(xs, ys):
        y_pred = forward(x)
        cost += (y_pred - y) ** 2
    return cost / len(xs)
    
```

```python
def gradient(xs, ys):
    grad = 0
    for x, y in zip(xs, ys):
        grad += 2 * x * (x * w - y)
    return grad / len(xs)
```

```python
print("predict (before training)", 4, forward(4))
epoces = range(100)
costs = []
for epoch in range(100):
    cost_val = cost(x_data, y_data)
    costs.append(cost_val)
    grad_val = gradient(x_data, y_data)
    w -= 0.01 * grad_val
    print("Epoch: ", epoch, "w=", w, 'loss=', cost_val)
print("predict (after training)", 4, forward(4))
```

```python
plt.plot(epoces, costs)
plt.ylabel('Cost')
plt.xlabel('epoces')
plt.show()
```
