```python
import torch
import matplotlib.pyplot as plt
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
```

```python
w = torch.Tensor([1.0])
w.requires_grad = True
```

```python
def forward(x):
    return x * w
```

```python
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) ** 2
    
```

```python
w = torch.Tensor([1.0])
w.requires_grad = True
print("predict (before training)", 4, forward(4).item())
epoces = range(300)
losts = []
for epoch in range(100):
    for x, y in zip(x_data, y_data):
        l = loss(x, y)
        l.backward()
        print("\tgrad: ", x, y, w.grad.item())
        w.data = w.data - 0.01 * w.grad.data
        losts.append(l.item())
        w.grad.data.zero_()
    print("Epoch: ", epoch, 'loss=', l.item())
print("predict (after training)", 4, forward(4).item())
```

```python
plt.plot(epoces, losts)
plt.ylabel('Loss')
plt.xlabel('epoces')
plt.show()
```
